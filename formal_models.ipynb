### 0.3 Visualization

There are 2 ways of structures representation:

#### 1. As a tuple notation: 

```python
print(my_structure)
```

#### 2. As a graph:

```python
my_structure.draw(title: str = None, 
                  dpi: int = 200, 
                  use_a_star: bool = True, 
                  engine: str = 'dot',
                  alignment: str = 'horizontal',
                  _dir: str = 'Graphs', 
                  _file: str = None, 
                  _show: bool = True,
                  _show_isolated_states: bool = False,
                  graph_kwargs: dict = None,
                  edge_kwargs: dict = None,
                  node_kwargs: dict = None,
                  initial_node_kwargs: dict = None,
                  final_node_kwargs: dict = None
)
```

- title: matplotlib axes title
- dpi: size of graph (resolution)
- use_a_star: allows to replace transition names with every possible symbol with a "*" for compactness
- engine: pygraphviz engine type (dot / circo / neato / fdp / sfdp)
- alighment: vertical / horizontal
- _dir: directory to store plots (optional)
- _file: file name for plot (optional)
- _show_isolated_states: plot states which are completely disconnected from initial states
- graph_kwargs / edge_kwargs / ... : optional dictionaries to pass to the pygraphviz Digraph

#### 3. As a Networkx graph (experimental, only for Graph() subclasses):

```python
my_structure.draw(title: str = None, 
                  seed: int = None, 
                  use_a_star: bool = True,
                  _show_isolated_states: bool = False
)
```
- title: matplotlib axes title
- seed: random seed to fix the output
- use_a_star: allows to replace transition names with every possible symbol with a "*" for compactness
- _show_isolated_states: plot states which are completely disconnected from initial states


*There might be minor differences in notation for particular classes*


## 1. fm.Automaton

### 1.1 Instance Creation

Main class for Finite Automatons (there is another class for Oracles).

fa = fm.Automaton(

- states: 

    - collection (preferably set) of states as Hashable objects (strings, integers etc.)

- initial_states: 

    - subset of states defined as initial states
    - if there are no initial states, still provide an empty collection
    - if there is a single initial state, it can be optionally provided as an object (not as a collection)

- final_states:

    - subset of states defined as final states
    - if there are no final states, still provide an empty collection
    - if there is a single final state, it can be optionally provided as an object (not as a collection)

- transitions:

    - collection (preferably set) of transitions with a certain form:
    - each transition is a 3-tuple defined as: (state_1, symbol, state_2)
    - symbol can be replaced by "*" if you mean that any symbol from the alphabet if valid here

- alphabet:

    - collection (preferably set) of symbols allowed. It is an optional argument. If alphabet is not provided, it will be restored from the transitions. However, if there are only "*" symbols in the transitions, restoration will not be possible.

- name:
        
    - name of the automaton. By default, it is named 'A'.
    
)

*Remark: symbols of the alphabet can be strings or integers;*

Automaton has a typical string representation of the form A = (S, I, Σ, T, F).

With states as integers and no explicit alphabet:

````python
from logex import formal_models as fm

automaton = fm.Automaton(states={1, 2, 3, 4},
                         initial_states={1, 3},
                         final_states={2, 4},
                         transitions={(1, 'open', 3), (1, 'open', 4), 
                                      (2, 'open', 1), (2, 'close', 2), 
                                      (3, 'close', 4), (3, 'close', 3),
                                      (4, 'close', 2), (4, 'close', 4)},
                         name="A")
print(automaton)
automaton.draw(dpi=200, engine='circo', alignment='vertical')
````

With states as strings and "*" notation:

````python
print(' With "*" symbols '.center(60,'-'))
automaton = fm.Automaton(states={'A','B', 'C', 'D'}, 
                         initial_states={'A', 'C'},
                         final_states={'B', 'D'}, 
                         transitions={('A', '*', 'C'), ('A', 'a', 'D'), 
                                      ('B', 'a', 'A'), ('B', 'b', 'B'), 
                                      ('C', 'b', 'D'), ('C', 'b', 'C'),
                                      ('D', 'b', 'B'), ('D', '*', 'D')}, 
                         alphabet={'a','b'},
                         name="A")
print(automaton)
automaton.draw(dpi=200, engine='circo', use_a_star=True)
````

With 1 initial state and no final states:

````python
print(' Without final states '.center(60,'-'))
automaton_ = fm.Automaton(states={'A','B', 'C', 'D'}, 
                         initial_states='A',
                         final_states=None, 
                         transitions={('A', 'a', 'C'), ('A', 'a', 'D'), 
                                      ('B', 'a', 'A'), ('B', 'b', 'B'), 
                                      ('C', 'b', 'D'), ('C', 'b', 'C'),
                                      ('D', 'b', 'B'), ('D', 'b', 'D')},
                         name="A")
print(automaton_)
automaton_.draw()
````

### 1.2 Methods and properties

#### Properties:

It is possible to check if automaton is deterministic or complete:

```python
>>> fa.is_complete
>>> fa.is_deterministic
>>> fa.is_power_automaton
```
Both return just a boolean answer True or False


#### Methods:

```python
>>> fa.is_complete_explain()
>>> fa.is_deterministic_explain()
```
Both return a boolean answer True or False, plus short explanation

#### Accepted Language

Once created automaton becomes a Callable object.
You can pass an arbitarary word and check if it is accepted (True) or not accepted (False):

```python
    >>> fa('bbb')
        True
    >>> fa('aaa')
        False
```

*deterministic and completeness methods and properties are common for all descendents of fm.Graph()*

Accepted Language:

````python
# Accepted Language
automaton.draw()
words = ['abbba', 'bbbaaa', 'abbaa', '𝜖', 'bbbaabbaab', 
         'bbbbbbbbbbbbbbbbbbbbbbbb', 'aab', 'bbbaabbaa']

# Without Path
print('\nWithout Path')
for word in words:
    print(word.ljust(30, '.') + str(automaton(word, return_path=False)))

print('\nWith Path')
# With Path
for word in words:
    print(word.ljust(30, '.') + str(automaton(word, return_path=True)))
````

Determinism and completeness:

````python
automaton.draw(dpi=100)

det = 'non-' if not automaton.is_deterministic else ''

automaton.is_deterministic_explain()

complete = 'in' if not automaton.is_complete else ''

automaton.is_complete_explain()

print(f'Automaton {automaton.name} is {det}deterministic and {complete}complete\n')
````

### 1.3 Conversions

<pre>   ==> Power Automaton (fm.Automaton)</pre>

Let's define 2 small automata:

````python
A = fm.Automaton(
    {'s0', 's1', 's2'},
    's0',
    's2',
    {('s0', 'a', 's1'), ('s0', 'a', 's2'), ('s1', 'b', 's2'), ('s2', '*', 's2')},
    name='A'
)
B = fm.Automaton(
    {'t0', 't1', 't2'},
    't0',
    't2',
    {('t0', '*', 't0'), ('t0', 'b', 't1'), ('t1', 'a', 't2')},
    name='B'
)

A.draw(dpi=100)
B.draw(dpi=100)
````

<pre>   ==> Power Automaton (fm.Automaton)</pre> 

````python
power_A = A.to_power_automaton()
power_A.draw()
````

<pre>   ==> Complement Automaton (fm.Automaton)</pre> 

````python
c_A = A.to_complement()
c_A.draw()

word = 'bba'
print(f'Word {word} accepted by A? \t {A(word)}')
print(f'Word {word} accepted by C(A)? \t {c_A(word)}')
````

<pre>   ==> Product Automaton (fm.Automaton)</pre> 

````python
AB = A.product(B)
AB.draw()
````

A acceptes words that start with "a...", B accepts words that end with "...ba".
A*B accpets words like "a ... ba"

````python
oracle = A.to_oracle()
oracle.draw()
````

<pre>   ==> Optimized Oracle (fm.Oracle)</pre> 

````python
opt_oracle = A.to_optimized_oracle()
opt_oracle.draw()
````

<pre>   ==> I/O Automaton (fm.IOAutomaton)</pre> 

(output of transitions quitting the final states are equal to 1)

````python
io = A.to_io_automaton()
io.draw()
````

(final states are lost)

````python
lts = A.to_lts()
lts.draw()
````

(from IO Automaton with unlabeled transitions)

````python
kripke = A.to_kripke(mode="s+i+o", linker='|') # see chapter about Kripke and IO Automaton
kripke.draw()
````

## 2. fm.Oracle

### 2.1 Instance Creation

Main class for Oracles

fa = Oracle(
- states: 
    - collection (preferably set) of states as Hashable objects (strings, integers etc.)

- initial_states: 
    - subset of states defined as initial states
    - if there are no initial states, still provide an empty collection
    - if there is a single initial state, it can be optionally provided as an object (not as a collection),

- final_states:
    - subset of states defined as final states
    - if there are no final states, still provide an empty collection
    - if there is a single final state, it can be optionally provided as an object (not as a collection)

- transitions:
    - collection (preferably set) of transitions with a certain form:
    - each transition is a 3-tuple defined as: (state_1, (symbol, index), state_2)
    - index can be an integer for Optimized Oracles or strings (or other Hashable object) for a simple Oracle (in the latter case, indices must be states)
    - Index can also be replaced by "*" if any index is valid.

- alphabet:
    - collection (preferably set) of 2-tuples allowed. It is an optional argument. If alphabet is not provided,
    - it will be restored from the transitions. As long as there might be very large alphabets for Oracles, using this argument is not recommended.

- name:
    - name of the oracle. By default, it is named 'A'.
    
        )

*Remark: symbols of the alphabet (first entries of tuples) can be strings or integers*

Automaton has a typical string representation of the form A = (S, I, Σ, T, F).

````python
from logex import formal_models as fm
oracle = fm.Oracle({'A','B', 'C', 'D'},
                    {'A'},
                    {'D'}, 
                    {('C', ('b', 'D'), 'D'), ('D', ('b', 'D'), 'D'), ('A', ('a', 'C'), 'C'), ('A', ('b', 'B'), 'B'),
	                            ('D', ('a', 'D'), 'D'), ('B', ('b', 'D'), 'D'), ('A', ('b', 'C'), 'C')}, 
                    name = "my_oracle")
print(oracle)
oracle.draw()
````

However, it is more typical to create Oracle from Automaton (see fm.Automaton, conversions)

```python
oracle.is_deterministic -> bool
oracle.is_complete -> bool
oracle.is_deterministic_explain() -> bool, print(str)
oracle.is_complete_explain() -> bool, print(str)
```

````python
print(oracle.is_deterministic)
oracle.is_deterministic_explain();
````

````python
print(oracle.is_complete)
oracle.is_complete_explain();
````

Accepted Language

The same as for fm.Automaton, using Oracle as Callable object
```python
oracle(word: Sequence, return_path: bool) -> bool
```

````python
# Accepted Language
oracle.draw()
words = ['abbba', 'bbbaaa', 'abbaa', '𝜖', 'bbbaabbaab', 
         'bbbbbbbbbbbbbbbbbbbbbbbb', 'aab', 'bbbaabbaa']

# Without Path
print('\nWithout Path')
for word in words:
    print(word.ljust(30, '.') + str(oracle(word, return_path=False)))

print('\nWith Path')
# With Path
for word in words:
    print(word.ljust(30, '.') + str(oracle(word, return_path=True)))
````

````python
oracle_ = fm.Oracle({'A','B'},
                    {'A'},
                    {'B'}, 
                    {('A', ('open', 'B'), 'B'), ('B', ('close', 'A'), 'A')}, 
                    name = "my_oracle_complex")
oracle_.draw()

print(f'Oracle {oracle_.name} accepts open-close? \t\t {oracle_(['open', 'close'])}')
print(f'Oracle {oracle_.name} accepts open-close-open? \t {oracle_(['open', 'close', 'open'])}')
````

## 3. fm.DigitalCircuit

### 3.1 Instance Creation

Helper class to describe digital circuits

circuit = fm.DigitalCircuit(

- inputs (Collection[str] | str, None): 

    - names of input channels, e.g. 'input' or ['in1', 'in2'], etc.

- outputs (Collection[str]):

    - names of output channels, e.g. 'output', ['out1', 'out2'], etc.
        
- states (Collection[str]):

    - names of internal circuit states, e.g. 'D', ['x', 'y'], etc.

- logical_rules (dict):

    - dictionary of a specific form: 
    - keys are names of states and output channels (if they are present)
    - values are functions that take exactly ONE argument of dict type and use the keys specified before to derive a logical step.
                
Example of logical_rules:
<pre>
{
    'state_1': lambda x: x['input'] and not x['state_2']
    'state_2': lambda x: not x['input']
    'output': lambda x: not x['state_1] or not x['state_2']
}
</pre>

)

*Remark: states are variables that can hold their value for the next iteration of the circuit. If their value is read as an output, there still has to be created a state and than an output channel equal to the state value*

Supposing there is a circuit:

````python
from PIL import Image
img = Image.open('circuit_1.png')
img
````

There in one input channel "in"

There is one output channel "out"

There are 2 internal states: "a" and "b"

Rules are:
```
a <- (not A) AND in
b <- a XOR (b AND in)
out <- (not b) AND (not a)
```

````python
from logex import formal_models as fm

circuit = fm.DigitalCircuit(
    inputs = {'in'},
    states = {'a', 'b'},
    outputs={'out'},
    logical_rules={
        'a': lambda x: not x['a'] and x['in'],
        'b': lambda x: x['a'] != (x['in'] and x['b']),
        'out': lambda x: not x['a'] and not x['b']
    }
)
print(circuit)
````

More tricky case:

````python
from PIL import Image
img_ = Image.open('circuit_2.png')
img_
````

There in one input channel "i"

There is one 2 output channels "o", "x"

There is 1 internal state: "o"

Rules are:
```
o <- not (i XOR o)
x <- not (i XOR o)
```

````python
circuit_ = fm.DigitalCircuit(
    inputs = {'i'},
    states = {'o'},
    outputs={'o', 'x'},
    logical_rules={
        'o': lambda x: not (x['i'] != x['o']),
        'x': lambda x: not (x['i'] != x['o'])
    }
)
print(circuit_)
````

## 4. fm.IOAutomaton

### 4.1 Instance Creation

io = fm.IOAutomaton(

- states (Collection[Union[int, str]]):

    - states of I/O Automaton as strings or integers

- initial_state (str): 

    - exactly one initial state

- transitions (Collection[tuple]): 

    - transitions as tuples of the form (state_1, (input, output), state_2)

- input_alphabet (Collection): 

    - mandatory input alphabet

- output_alphabet (Collection): 

    - mandatory output alphabet

- name (str, optional): 

    - a name for I/O Automaton. Defaults to None.
)

*Remark: typically is cerated from FA automaton or from Digital Circuit (see above)*

````python
from logex import formal_models as fm

io = fm.IOAutomaton(

    states = {'s0', 's1'},
    initial_state='s0',
    input_alphabet={'a'},
    output_alphabet={'o', 'e'},
    transitions={
        ('s0', ('a', 'o'), 's1'), ('s1', ('a', 'e'), 's0')
    }
)
print(io)
io.draw()
````

### 4.2 Methods

A single method:
<pre>__call__</pre>

When called, O/I Automaton returns a sequnece of outputs produced by the sequence of inputs, or False, if input is not accepted.

```python
io(word: Sequnce, return_path: bool) -> Sequence | bool
```

````python
for i in range(1, 5):
    print('Input = ' + ('a' * i) + ' Output = ' + io('a' * i, return_path=False))

for i in range(1, 5):
    print('Input = ' + ('a' * i) + ' Output = ' + ' : '.join(io('a' * i, return_path=True)))
````

<pre>   ==> Finite Automaton (fm.Automaton)</pre>

(about Kripke - see below)

<pre>   ==> Finite AUtomaton (fm.Automaton)</pre> 

target_outputs: output(s) to treat like a target (by default, "1"). Target output does not mean the last symbol of accepted word but the auxilary symbol appended to it. States with outcoming target output transition will be called final states.

````python
A = io.to_finite_automaton(target_outputs=['o'])
A.draw()
````

<pre>   ==> Kripke (fm.Kripke)</pre>

linker: symbol to connect inputs, states and outputs with

mode: order of inputs/states/outputs in Kripke state names

no_annotations: hide annotations in Kripke

no_initial_states: make Kripke with nominitial states

````python
kripke = io.to_kripke(linker='|', mode='i+s+o', no_annotations=True)
kripke.draw()
````

## 5. fm.LabaledTransitionSystem (LTS)

### 5.1 Instance Creation

The simplest extension of fm.Graph() super-class. Just a collection of states with possibly
initial states and NO final states. 
Transitions connect pairs of states through arbitrary actions.

lts = fm.fm.LabaledTransitionSystem(

- states (Collection[Union[int, str]]): 

    - collection of states as integers or strings

- initial_states (Collection[Union[int, str]] | str | None): 

    - initial states (optional)

- transitions (Collection[tuple]): 

    - collection of 3-tuple transitions (state_1, action, state_2)

- actions (Collection, optional): 

    - collection of actions. Can be obtained from transitions. Defaults to None.

- name (str, optional): 

    - name for LTS. Defaults to None.
)

*Remark: Automaton without final states is actually a LTS*

## 8. fm.ProcessAlgebra

### 8.1 Instance Creation

A special class for holding process algebra systems

pral = fm.ProcessAlgebra(rules)

Rules must be a collection of strings of the form: "Process1 = expression1"
By syntax, processes must start with uppercase letter, while actions - with lowercase.

Apart from '.' connector, there is a "+" connector allowed.

Also, parallel operator "||" can be used to join several independent processes together.

Examples of valid rules:

- "P = a.c.P"
- "R = e.R"
- "Q = a.c.f.P + b.f.Q"
- "T = q.p.(f.P + b.P)"
- "Parallel = P || R"

````python
from logex import formal_models as fm

pral = fm.ProcessAlgebra(
    [
        'P = a.(c.P + d.P)',
        'Q = d.c.f.Q',
        'R = e.c.f.R',
        'Parallel = P || Q || R'
    ]
)
print(pral)
````

### 8.2 Methods

There is a single special method:
```python
.apply_action_to_process(process: str, action: str) -> list
```

Returns all possible processes resulting from the application

````python
pral.apply_action_to_process(process='P', action='a')
````

````python
pral.apply_action_to_process(process='P || Q', action='a')
````

### 8.3 Conversions

<pre>   ==> LTS (fm.LabeledTransitionSystem)</pre> 

In case of a single process:

````python
lts = pral.to_lts(initial='P')
lts.draw(dpi=100)
lts = pral.to_lts(initial='Q')
lts.draw(dpi=100)
lts = pral.to_lts(initial='R')
lts.draw(dpi=100)
````

In case of parallel operator:

````python
lts = pral.to_lts(initial='Parallel')
lts.draw(alignment='vertical')
````

## 9. fm.Kripke

### 9.1 Instance Creation

Kripke structure instance, sub-class of Graph() but with no transition labels 
and sates annotated with propositional variables

K = fm.Kripke(

- states (Collection[Union[int, str]]): 

    - states as a collection

- initial_states (Collection[Union[int, str]] | str | None): 

    - initial states

- transitions (Collection[tuple]):

    - transitions as 2-tuples (state_1, state_2)

- annotations (dict[Union[str, dict, Collection]], optional): 

    - annotations as a dict. Defaults to None.

- name (str, optional): 

    - name for Kripke Structure. Defaults to None.

)

Annotations should be a dictionary if provided. Keys - sates, values - propositional variables, whcih are true in this state. If only one variable is true in this state - can be just a string with this variable.
Also value can be an expicit dictionary with propositional values:

```python
annotations = {'B': 'p', 'C': 'q', 'D': ['p', 'q']}

annotations = {'B': ['p'], 'C': ['q'], 'D': ['p', 'q']}

annotations = {'B': {'p': 1, 'q': 0}, 'C': {'q': 1, 'p': 0}, 'D': ['p', 'q']}
```

All are the same

````python
from logex import formal_models as fm

K = fm.Kripke(
    states = {'A', 'B', 'C', 'D'},
    initial_states = 'A',
    transitions = {('A', 'C'), ('C', 'A'), ('A', 'B'),
     ('B', 'A'), ('B', 'D'), ('D', 'A')},
    annotations = {'B': 'p', 'C': 'q', 'D': ['p', 'q']},
    name = 'K'
)
print(K)
K.draw()
````
